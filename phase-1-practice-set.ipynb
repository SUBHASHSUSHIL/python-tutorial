### Joining Sets

Python provides multiple ways to join two or more sets:

1. union() – Returns a new set containing all unique elements from both sets.

2. intersection() & update() – Adds elements from one set to another, modifying the original set.

3. intersection_update() – Keep only the common elements.

4. symmetric_difference() & symmetric_difference_update() – Keep elements that are unique to each set.

## Joining Two Sets

### Using union()

```python
# Define two sets
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision"}
big_data_topics = {"Hadoop", "Spark", "Kafka", 1, 2}

# Create a new set with all unique elements from both sets
all_topics = ai_topics.union(big_data_topics)

# Display the result
print("All Topics Set:", all_topics)
```

### Using the | operator

```python
# Define two sets using `|` set unique operator
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision"}
data_topics = {"big data", "data engineering", "data analysis", 1, 2}

# Create a new set with all unique elements
combined_topics = ai_topics | data_topics

# Display the result
print("Combined Topics Set:", combined_topics)
```

### Using update()

```python
# Define two sets
set1 = {"python", "ai", "ml", "data science", "data engineering"}
set2 = {"machine learning", "deep learning", "computer vision", "nlp", "data analysis"}

# Update set1 with elements from set2
set1.update(set2)

# Display the updated set
print("Updated Set1 with Set2 elements:", set1)
```

## Keeping Only the Common Elements
The intersection_update() method keeps only elements present in both sets, modifying the original set.

The intersection() method returns a new set containing only common elements.

### Using intersection_update()

```python
# Define two sets
set_a = {"Python", "Java", "C++", "JavaScript"}
set_b = {"Java", "C#", "Ruby", "JavaScript", "Go"}

# keep only common elements
set_a.intersection_update(set_b)

# Display the updated set
print("Set A after intersection update with Set B:", set_a)
```

### Using intersection()

```python
# Define two sets
set_a = {"Python", "Java", "C++", "JavaScript"}
set_b = {"Java", "C#", "Ruby", "JavaScript", "Go"}

# Create a new set with elements in set_a but not in set_b
common_set = set_a.intersection(set_b)

# Display the result
print("Common elements between Set A and Set B:", common_set)
```

### Using set intersection (&) operator

```python
# Define two sets using `g` operater

setA = {"Python", "Java", "C++", "JavaScript"}
setB = {"Java", "C#", "Ruby", "JavaScript", "Go"}

# Create a new set with common elements
commonSet = setA & setB

# Display the result
print("Common elements between Set A and Set B using '&' operator:", commonSet)
```

## Keeping All, But Not the Common Elements

The **symmetric_difference_update()** method removes common elements, keeping only unique ones.

The **symmetric_difference()** method returns a new set with only unique elements from both sets.

### Using symmetric_difference_update()

```python
# Define two sets
set_x = {"Data Science", "Machine Learning", "AI", "Deep Learning"}
set_y = {"AI", "Blockchain", "Cybersecurity", "Data Science", "IoT"}

# Keep only non-common elements
set_x.symmetric_difference_update(set_y)

# Display the updated set
print("Set X after symmetric difference update with Set Y:", set_x)
```

### Using symmetric_difference()

```python
# Define two sets
set_a = {"Python", "Java", "C++", "JavaScript"}
set_b = {"Java", "C#", "Ruby", "JavaScript", "Go"}

# Create a new set with only non-common elements
unique_set = set_a.symmetric_difference(set_b)

# Display the result
print("Unique elements between Set A and Set B:", unique_set)
```

# Python Set Methods

1.    **add()**                          Inserts an element into the set.

2.    **clear()**                        Removes all elements, leaving the set empty.

3.    **copy()**                         Creates and returns a duplicate of the set.

4.    **difference()**                   Returns a new set containing elements that exist in the first set but not in others.

5.    **difference_update()**            Modifies the set by removing elements that are also found in another set.

6.    **discard()**                      Eliminates the specified element if it exists; does nothing if the element is absent.

7.    **intersection()**                 Generates a new set containing only elements that are common in multiple sets.

8.    **intersection_update()**          Updates the set, keeping only elements found in all specified sets.

9.    **isdisjoint()**                   Checks whether two sets share no common elements, returning True or False.

10.   **issubset()**                     Determines if the current set is completely contained within another set.

11.   **issuperset()**                   Verifies whether the current set includes all elements of another set.

12.   **pop()**                          Removes and returns an arbitrary element from the set.

13.   **remove()**                       Deletes a specified element from the set; raises an error if the element is missing.

14.   **symmetric_difference()**         Returns a new set with elements unique to each set, excluding common elements.

15.   **symmetric_difference_update()**  Updates the set by keeping only elements that are not present in both sets.

16.   **union()**                        Returns a set containing all elements from both sets, eliminating duplicates.

17.   **update()**                       Merges another set into the current one, adding any new elements.

## 1. add()

The **add()** method inserts a new element into a set. If the element already exists, the set remains unchanged.

Syntax:- **set.add(element)**

Parameter Values:- **element:** Required. The element to add to the set

### Useful Information

**No Duplicate Insertions:-**  If the element is already present in the set, it is not added again.

**No Return Value:-**  The method modifies the set in place and does not return a new set.

**Mutable but Unordered:-***  The insertion order is not preserved in sets.

```python
# Creating a set of programming topics covered on intensity coding platform
programming_topics = {"Python", "Java", "C++", "JavaScript", "Data Science", "Machine Learning", "AI", "Web Development", "Mobile Development", "Cloud Computing"}

# Adding a new topic to the set
programming_topics.add("DevOps")

print("Updated Programming Topics Set:", programming_topics)

programming_topics.add("Java")

print("Programming Topics Set after adding 'Java' again:", programming_topics)
```

## 4. difference()

The **difference()** method returns a new set containing elements that are in the first set but not in the second set.

Syntax:-  **set1.difference(set2)**

Parameter Values:-  **set2**: Required. The set whose elements are excluded from the result

```python
# Topics covered on intensity coding platform  
topics_covered = {"Python", "Java", "C++", "JavaScript", "Data Science", "Machine Learning", "AI", "Web Development", "Mobile Development", "Cloud Computing"}

# Topics covered on another platform  
other_platform_topics = {"Python", "Ruby", "Go", "JavaScript", "Data Analysis", "Cybersecurity", "AI", "DevOps"}

# Finding topics unique to intensity coding platform
unique_topics = topics_covered.difference(other_platform_topics)
print("Topics unique to Intensity Coding platform:", unique_topics)

# Finding topics unique to other platform
unique_other_topics = other_platform_topics.difference(topics_covered)
print("Topics unique to Other platform:", unique_other_topics)
```

##  5. difference_update()

The **difference_update()** method removes elements from the first set that are also present in the second set. Unlike **difference()**, this method modifies the riginal set instead of returning a new one.

Syntax:-  **set1.difference_update(set2)**

Parameter Values:-  **set2**: Required. The set whose common elements will be removed from set1.


```python
# Topics covered on intensity coding platform  
topics_covered = {"Python", "Java", "C++", "JavaScript", "Data Science", "Machine Learning", "AI", "Web Development", "Mobile Development", "Cloud Computing"}

# Topics covered on another platform
other_platform_topics = {"Python", "Ruby", "Go", "JavaScript", "Data Analysis", "Cybersecurity", "AI", "DevOps"}

# Removing topics that exist on both platforms
topics_covered.difference_update(other_platform_topics)

print("Topics covered on Intensity Coding platform after difference update:", topics_covered)
```

## 7. intersection()

The **intersection()**  method returns a set containing only the elements that are common between two or more sets.

Syntax:- **set.intersection(set1, set2, ...)**

Parameter Values:- **set1**: Required. The first set to compare.

                   **set2, ...**: Optional. Additional sets to compare. Multiple sets can be provided, separated by commas

```python
# AI topics disscussed on different forums
forum1_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision"}
research_papers_topics = {"Reinforcement Learning", "GANs", "NLP", "Computer Vision"}
industry_trends_topics = {"AI Ethics", "AI in Healthcare", "NLP", "Computer Vision"}

# Finding topics common between Intensity Coding and research papers
common_with_research = forum1_topics.intersection(research_papers_topics)

print("Common topics between Forum 1 and Research Papers:", common_with_research)

common_all = forum1_topics.intersection(research_papers_topics, industry_trends_topics)
print("Common topics across all three sources:", common_all)
```

## 8. intersection_update()

The **intersection_update()** method removes elements from the original set that are not present in another set or multiple sets.

Syntax:- **set.intersection_update(set1, set2, ...)**

Parameter Values:- **set1**: Required. The first set to compare.

                   **set2, ...** Optional. Additional sets to compare. Multiple sets can be provided, separated by commas

```python
# AI topics covered on Intensity Coding platform
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

research_topicss = {"NLP", "Computer Vision", "AI Ethics", "GANs"}

# Keep only the topics that are present in both sets
ai_topics.intersection_update(research_topicss)
print("AI Topics after intersection update with Research Topics:", ai_topics)

x = {"A", "B", "C", "D"}
y = {"C", "D", "E", "F"}
z = {"D", "E", "F", "G"}
x.intersection_update(y, z)
print("X after intersection update with Y and Z:", x)
```

## 9. isdisjoint()

The **isdisjoint()** method checks whether two sets have no common elements. If the sets have no intersection, it returns **True**; otherwise, it returns False.

Syntax:- **set.isdisjoint(other_set)**

Parameter Value:- **other_set**:  Required. The set to compare for common elements


```python
# AI topics covered on Intensity Coding platform  
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

# topics covered cyber security platform
cyber_security_topics = {"Network Security", "Cryptography", "Ethical Hacking", "Malware Analysis"}

# Check if there is no overlap
no_overlap = ai_topics.isdisjoint(cyber_security_topics)

print("Is there no overlap between AI topics and Cyber Security topics?", no_overlap)

# Another example where set share an elements
other_ai_topics = {"NLP", "Computer Vision", "AI Ethics", "GANs"}
no_overlap = ai_topics.isdisjoint(other_ai_topics)
print("Is there no overlap between AI topics and Other AI topics?", no_overlap)
```

## 10. issubset()

The **issubset()** method checks whether all elements of a set exist in another set. It returns **True**  if the set is a subset, otherwise **False**.

Syntax:- **set.issubset(other_set)**

Parameter Values:- **other_set** : Required. The set to compare against


```python
# Core AI topics covered on Intensity Coding platform
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

# All topics covered including additional ai topics
all_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning", "AI Ethics", "GANs"}

# Check if all core topics are part of the full topic list
is_subset = ai_topics.issubset(all_topics)

print("Are all AI topics a subset of all topics?", is_subset)

# Example where a set is not a subset
random_topics = {"Blockchain", "Cybersecurity", "IoT"}

# Check if random_topics is a subset of all_topics
is_subset_random = random_topics.issubset(all_topics)
print("Are Random Topics a subset of All Topics?", is_subset_random)
```

## 11. issuperset()

The **issuperset()** method checks whether all elements of another set exist in the current set. It returns **True**
 if the current set is a superset, otherwise **False**.

Syntax:- **set.issuperset(other_set)**

Parameter Values:- **other_set**: Required. The set to compare against

```python
# All ai topics covered on Intensity Coding platform
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

# Ai core topics
core_ai_topics = {"Machine Learning", "Deep Learning", "NLP"}

# Check if core_ai_topics covers all core topics
is_superset = ai_topics.issuperset(core_ai_topics)
print("Does AI topics set superset core AI topics?", is_superset)

# Another example where set is not a superset
additional_topics = {"Machine Learning", "Deep Learning", "NLP", "AI Ethics", "GANs", "Blockchain"}
is_superset_additional = ai_topics.issuperset(additional_topics)
print("Does AI topics set superset additional topics?", is_superset_additional)
```

## 14. symmetric_difference()

The **symmetric_difference()** method returns a set containing elements that are in either of the two sets but not in both.

Syntax:- **set.symmetric_difference(other_set)**

Parameter Values:- **other_set**: Required. The set to compare against

```python
# AI topics covered on Intensity Coding platform  
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

# Ai topics covered on another platform  
cyber_security_topics = {"Network Security", "Cryptography", "Ethical Hacking", "Malware Analysis"}

# Find unique topics in both platforms
unique_ai_topics = ai_topics.symmetric_difference(cyber_security_topics)
print("Unique AI Topics between Intensity Coding and Cyber Security platforms:", unique_ai_topics)
```

## 15. symmetric_difference_update()

The **symmetric_difference_update()** method updates the original set by removing elements that exist in both sets and adding elements that
are unique to either set.

Syntax:- **set.symmetric_difference_update(other_set)**

Parameter Values:- **other_set**: Required. The set to compare against.


```python
# AI topics covered on Intensity Coding platform  
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

# Ai topics covered on another platform
cyber_security_topics = {"Network Security", "Cryptography", "Ethical Hacking", "Malware Analysis","NLP", "Computer Vision"}

# update ai_topics topics to include unique topics while removing common ones
ai_topics.symmetric_difference_update(cyber_security_topics)
print("AI Topics Set after symmetric difference update with Cyber Security topics:", ai_topics)
```

## 16. union()

The **union()** method creates a new set containing all unique elements from the original set and one or more specified sets (or iterables).

Syntax:- **set.union(set1, set2, ...)**

Parameter Values:- **set1, set2, ...**: Required. The sets or iterables to combine

```python
# AI topics covered on Intensity Coding platform  
ai_topics = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision", "Reinforcement Learning"}

# Another platform covered topics in ai and cyber security
another_platforms = {"Network Security", "Cryptography", "Ethical Hacking", "Malware Analysis", "NLP", "Computer Vision"}

# topics cover in research instititutes
research_institute_topics = {"AI Ethics", "GANs", "Reinforcement Learning", "Computer Vision"}

# Creating a union of all ai topics
all_topics = ai_topics.union(another_platforms, research_institute_topics)

print("All AI Topics from different sources:", all_topics)
```
