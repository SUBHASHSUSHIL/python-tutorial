### Working with Different Data Types

```python
# Sets containing same data type
coding_set1 = {"Python", "Java", "C++"}  #string
coding_set2 = {10, 40, 20, 30}  #integer
coding_set3 = {True, False, False, True}  #boolean

# Set with mixed data types
coding_mixed_set = {"Intensity Coding", 10, True, "AI", 20.5}

# Displaying the sets
print("Set with strings:", coding_set1)
print("Set with integers:", coding_set2)
print("Set with booleans:", coding_set3)
print("Set with mixed data types:", coding_mixed_set)
```

### Adding Items from Another Sets

```python
# Define two sets of programming languages
ai_set = {"Python", "Java", "C++", "JavaScript"}
new_tech = {"Go", "Rust", "Kotlin", "Swift"}

# add new_tech into ai_set
ai_set.update(new_tech)

# Display the updated ai_set
print("Updated AI Set with New Technologies:", ai_set)
```

### Removing Items from a Set

Python provides multiple ways to remove elements or clear an entire set, each with distinct
behaviors. Let’s explore them one by one.

remove() – Removes a specific item.
Raises an error if the item is not found.

discard() – Removes a specific item.
Does not raise an error if the item is not found.

pop() – Removes a random item.
Since sets are unordered, you cannot predict which item will be removed.
Returns the removed item.

clear() – Empties the set, removing all elements.

del keyword – Deletes the set entirely.
